```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
filenames = 'cbook2.png'
img_path = f"images/{filenames}"
img = cv2.imread(img_path)
```

```python
# creating data of right size 
def Display(im_path):

    dpi = 80
    im_data = plt.imread(im_path)
    if len(im_data.shape) == 3:
        height, width, depth = im_data.shape  # Color image (height, width, depth)
    else:
        height, width = im_data.shape  # Grayscale image (height, width)
        depth = None 

    # What size does the figure need to be in inches to fit the image?
    figsize = width / float(dpi), height / float(dpi)

    # Create a figure of the right size with one axes that takes up the full figure
    fig = plt.figure(figsize=figsize)
    ax = fig.add_axes([0, 0, 1, 1])

    # Hide spines, ticks, etc.
    ax.axis('off')

    # Display the image.
    ax.imshow(im_data, cmap='gray')

    plt.show()
```

```python
Display(img_path)
```

## binarization 

```python
# this meanse Greyscaling the image like increasing hue and all
def binarization(img):
    return cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
```

```python
grey_image = binarization(img)
cv2.imwrite(f"temp/grey{filenames}" , grey_image)
```

```python
Display(f"temp/grey{filenames}")
```

## Decreasing the thresh

```python
thresh ,im_bw = cv2.threshold(grey_image,210,230,cv2.THRESH_BINARY)
cv2.imwrite(f"temp/bw{filenames}" , grey_image)
```

```python
Display(f"temp/bw{filenames}")
```

# Noice Removal

```python
def noise_removal(image):
    kernel = np.ones((1,1) , np.uint8)
    image = cv2.dilate(image, kernel , iterations=1)
    kernel = np.ones((1,1) , np.uint8)
    image = cv2.erode(image , kernel , iterations=1)
    image = cv2.morphologyEx(image , cv2.MORPH_CLOSE , kernel)
    image = cv2.medianBlur(image , 3)
    return (image) 
```

```python
no_noice = noise_removal(im_bw)
cv2.imwrite(f"temp/no_noise{filenames}" , no_noice)
```

```python
Display(f"temp/no_noise{filenames}")
```
